```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'normal':np.random.normal(size = 60),
                   'gamma':np.random.gamma(1,size = 60)})
df.plot()
```

```python
stu = {'name':['小程','李四','张三','小鱼','奇奇'],
'sex':['male','male','male','female','female'],
'year':[2004,2005,2003,2004,2003]}
data = pd.DataFrame(stu)
print(data['sex'].value_counts())
data['sex'].value_counts().plot(kind = 'bar',rot = 30)
```

```python
data = np.random.normal(loc=5, scale=2, size=2000)  
plt.hist(data, bins=40, color='lightgreen', edgecolor='darkgreen', alpha=0.65)
plt.title('Modified Histogram') 
plt.xlabel('Measurement')       
plt.ylabel('Occurrences')      
plt.show()
```

```python
x = np.random.rand(80) * 15 
y = x * 3 + np.random.randn(80) * 2.5 
plt.scatter(x, y, c='orange', edgecolor='darkred', s=70, alpha=0.6)
plt.title('Modified Scatter Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True, linestyle=':', alpha=0.4)
plt.show()
```
